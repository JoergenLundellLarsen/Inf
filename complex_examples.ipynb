```python
from complex import Complex

z = Complex(1, 2)
y = Complex(3, 4)
print(z)
print([z,y])
print(z.real,z.imag)
print(Complex())
print(Complex(5))
print(z * y)
print(z + y)
print(z - y)

print(z + 3)    # 4+2i
print(3 + z)    # 4+2i
print(z * 3)    # 3+6i
print(3 * z)    # 3+6i
print(z - 3)    # -2+2i
print(3 - z)    # 2-2i

print(z==y)
print(z!=y)
```
